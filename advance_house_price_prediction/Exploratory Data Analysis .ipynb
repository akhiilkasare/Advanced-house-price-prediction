# 1. Missing Values

```python
# step 1-Listing the features which has missing values
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

#Printing feature name with % of missing values
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4), '% of missing values')
```

# Since there are many missing values we need to check the relation between 
# the missing values and Sales Price(Dependent feature)

```python
for feature in features_with_na:
    data = dataset.copy()
    
    #Making a variable if the data is missing it will show 1 otherwise it'll show 0
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    #Calculating the mean sale price where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

With the relation between the missing values and the dependent features we need to replace nan values with
something meaningful

# Numerical Variables

```python
#Listing all the numerical variable from the dataset
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype !='O']
print("The length of the numerical features is : ", len(numerical_features))

dataset[numerical_features].head()
```

```python
# List of variables which contains year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
#Checking whether there is a relation between which year the house is sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

Numerical variable are basically of two types: 
1. Continous variables 2. Discrete Variables

```python
#we are selecting 25 as a threshold value means it should not be greater than 25
#and the numerical data should not be in year and ID column
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('Discrete feature variable count : ', len(discrete_feature))
```

- Exploratory Data Analysis is to gather information from each and every feature and you need to compare each and 
  every Independent feature with the Dependent feature

```python
dataset[discrete_feature].head()
```

```python
# Finding the relationship between discrete features and Dependent feature

for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

# Continous Variable

```python
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print('The count of continous feature is : {}'.format(len(continous_feature)))
```

```python
for feature in continous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

As the graphs are skewed we need to logarithmic transformation 

```python
for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
```

# Checking Outliers

Outliers basically means suppose in a distribution there are values which are completely outside the a particular way
whether it has a very high value or very low value those kind of values are basically called as outliers

```python
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
